```python
import pandas as pd
```

```python
df = pd.read_csv('stock_data.csv')
df
```

```python
#to skip the first line
df = pd.read_csv('stock_data.csv',skiprows=1) # OR use header=1
df
```

```python
#when there is no header
df = pd.read_csv('stock_data.csv',header=None)
df
```

```python
#to specify col name
df = pd.read_csv('stock_data.csv',header=None,names=['name','place','animal','thing']) 
df
```

```python
df = pd.read_csv('stock_data.csv')
df
```

```python
#read only 3 rows
df = pd.read_csv('stock_data.csv',nrows=3)
df
```

```python
#treat null values
df = pd.read_csv('stock_data.csv',na_values=['not available','n.a.'])
df
```

```python
#treat only nulls of specific cols (eps could be -1 but not the revenue)
df = pd.read_csv('stock_data.csv',na_values={
    'eps':['not available','n.a.'],
'revenue':['not available','n.a.',-1],
'price':['not available','n.a.'],
'people':['not available','n.a.'],})
df
```

```python
#Write to csv
df.to_csv('clean.csv',index=False)
```

```python
#write only 2 cols
df.to_csv('clean.csv',columns=['tickers','eps'],index=False)
```

```python
#skip header
df.to_csv('clean.csv',header=False,index=False)
```
