```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.axis="equal"
plt.pie(exp_vals,labels=exp_labels,radius=0.75,autopct='%0.2f%%') #for two decimal points
plt.show()
```

```python
plt.axis="equal"
plt.pie(exp_vals,labels=exp_labels,radius=0.75,autopct='%0.2f%%',explode=[0,0.25,0,0,0]) #for two decimal points
plt.show()
```

```python
plt.axis="equal"
plt.pie(exp_vals,labels=exp_labels,radius=0.75,autopct='%0.2f%%',startangle=45) #for two decimal points
plt.show()
```
