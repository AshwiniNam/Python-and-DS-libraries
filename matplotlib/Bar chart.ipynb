```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit = [40,2,34,12]
```

```python
com = np.arange(len(company))
com
```

```python
#plt.xticks(com,company)
#plt.bar(com,revenue)

plt.xticks(com,company)
plt.bar(com-0.1,revenue,label='Revenue',width=0.2) #this can be done only to a np array
plt.bar(com+0.1,profit,label='Profit',color='Green',width=0.2)
plt.legend()
```

```python
#horizontal chart
plt.yticks(com,company)
plt.barh(com-0.1,revenue,label='Revenue',height=0.2) #this can be done only to a np array
plt.barh(com+0.1,profit,label='Profit',color='Green',height=0.2)
plt.legend()
```
